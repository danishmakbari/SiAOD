```python
# 1. –ò–º–µ–µ—Ç —Å–º—ã—Å–ª –±—Ä–∞—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤—ã —Ç–æ–ª—å–∫–æ —á–µ—Ç–Ω–æ–π –¥–ª–∏–Ω—ã, —Ç.–∫. —Ä–∞–∑–≤–æ—Ä–æ—Ç –Ω–µ—á–µ—Ç–Ω—ã—Ö –Ω–µ –¥–∞–µ—Ç –Ω–∏–∫–∞–∫–æ–≥–æ —ç—Ñ—Ñ–µ–∫—Ç–∞.
# 2. –†–∞–∑–≤–æ—Ä–æ—Ç –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ —á–µ—Ç–Ω–æ–π –¥–ª–∏–Ω—ã –¥–µ–ª–∞–µ—Ç –≤—Å–µ —á–µ—Ç–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã –Ω–µ—á–µ—Ç–Ω—ã–º–∏ –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç.
# 3. –ß—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –∫–∞–∫–æ–π –ø–æ–¥–º–∞—Å—Å–∏–≤ —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—å, –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ shift –≤ –∫–æ—Ç–æ—Ä–æ–º –∏–∑ –∫–∞–∂–¥–æ–≥–æ –Ω–µ—á–µ—Ç–Ω–æ–≥–æ
# –≤—ã—á—Ç–µ–º —Å–æ—Å–µ–¥–Ω–µ–µ —á–µ—Ç–Ω–æ–µ. –≠—Ç–æ –Ω–∞–º –¥–∞—Å—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø—Ä–∏—Ä–æ—Å—Ç–µ —Å—É–º–º—ã. –¢–∞–∫–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤ shift –±—É–¥–µ—Ç –¥–≤–∞.
# –ü–µ—Ä–≤—ã–π, –≥–¥–µ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å 1 —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∏–Ω—É—Å 0 —ç–ª–µ–º–µ–Ω—Ç–∞, –≤—Ç–æ—Ä–æ–π - —Å 1 —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∏–Ω—É—Å 2 —ç–ª–µ–º–µ–Ω—Ç,
# —Ç–∫ —Ä–∞–∑–≤–æ—Ä–∞—á–∏–≤–∞—Ç—å –º–∞—Å—Å–∏–≤ –º–æ–∂–Ω–æ –ª–∏–±–æ —Å —á–µ—Ç–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –ª–∏–±–æ –Ω–µ—á–µ—Ç–Ω–æ–≥–æ
# 4. –ò—â–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–º–∞—Å—Å–∏–≤ –≤ –¥–≤—É—Ö –º–∞—Å—Å–∏–≤–∞—Ö shift —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –∫–∞–∫–æ–π —Ä–∞–∑–≤–æ—Ä–æ—Ç –¥–∞—Å—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç.
# 5. –ò–Ω–¥–µ–∫—Å—ã –∏–∑ shift –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ –æ–±—ã—á–Ω—ã–π –º–∞—Å—Å–∏–≤
# 6. –†–∞–∑–≤–æ—Ä–∞—á–∏–≤–∞–µ–º –∏ —Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É. –ò–∑ –¥–≤—É—Ö —Å—É–º –≤—ã–±–∏—Ä–∞–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é

from sys import maxsize

def maxSubArray(a):
    size = len(a)
    max_so_far = -maxsize - 1
    max_ending_here = 0
    start = 0
    end = 0
    s = 0
    for i in range(0,size):
        max_ending_here += a[i]
        if max_so_far < max_ending_here:
            max_so_far = max_ending_here
            start = s
            end = i
        if max_ending_here < 0:
            max_ending_here = 0
            s = i+1
    if max_so_far > 0:
        return start, end
    else:
        return 0, -1

def task3(arr):
    shift = []
    for i in range(1, len(arr), 2):
        shift.append(arr[i] - arr[i - 1])
    start, end = maxSubArray(shift)
    start *= 2
    end = end * 2 + 1
    l = arr[:start]
    mid = arr[start:end + 1]
    mid.reverse()
    r = arr[end + 1:]
    l.extend(mid)
    l.extend(r)
    sum1 = 0
    for i in range(0, len(l), 2):
        sum1 += l[i]
    
    shift = []
    for i in range(2, len(arr), 2):
        shift.append(arr[i - 1] - arr[i])
    start, end = maxSubArray(shift)
    start = start * 2 + 1
    end = end * 2 + 1 + 1
    l = arr[:start]
    mid = arr[start:end + 1]
    mid.reverse()
    r = arr[end + 1:]
    l.extend(mid)
    l.extend(r)
    sum2 = 0
    for i in range(0, len(l), 2):
        sum2 += l[i]
    return max(sum1, sum2)

res = []
t = int(input())
for i in range(t):
    n = int(input())
    arr = input().split()
    arr = [int(arr[i]) for i in range(len(arr))]
    res.append(task3(arr))
for i in range(t):
    print(res[i])
```

```python
# –ê–ª–≥–æ—Ä–∏—Ç–º –ï–≤–∫–ª–∏–¥–∞
def gcd(a, b):
    while a != 0 and b != 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    return a + b

# –ù–û–î –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞
def gcd_arr(arr):
    result = arr[0]
    for i in range(1, len(arr)):
        result = gcd(result, arr[i])
    return result

def task7(a, b):
    result = []
    for j in range(len(b)):
        temp = []
        for i in range(len(a)):
            temp.append(b[j] + a[i])
        result.append(gcd_arr(temp))
    return result

[n, m] = input().split()
a = input().split()
a = [int(a[i]) for i in range(len(a))]
b = input().split()
b = [int(b[i]) for i in range(len(b))]
result = task7(a, b)
for item in result:
    print(item, end = ' ')
```

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤ [3,5]. –¢–æ–≥–¥–∞ ùëé=[1,8,3,6,7,6] –∏ 1‚ãÖ5+8‚ãÖ9+3‚ãÖ6+6‚ãÖ8+7‚ãÖ8+6‚ãÖ6=235.

```python
# –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ —Ü–µ–Ω—Ç—Ä—ã –ø–æ–¥–º–∞—Å—Å–∏–≤–æ–≤
# –ü–æ–¥–º–∞—Å—Å–∏–≤—ã –º–æ–≥—É—Ç –±—ã—Ç—å –∏–∑ —á–µ—Ç–Ω–æ–≥–æ –ª–∏–±–æ –Ω–∏—á–µ—Ç–Ω–æ–≥–æ –∫–æ–ª-–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
# –ù–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –ø—Ä–∏–±–∞–≤–ª—è–µ–º –∫ curr_max a[l] * b[l] + a[r] * b[r]
# –ê –∫ curr_rmax a[r] * b[l] + a[l] * b[r]
# –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É curr_rmax –∏ curr_max –ø–æ–∫–∞–∂–µ—Ç –Ω–∞–º –ø—Ä–∏—Ä–æ—Å—Ç –æ—Ç —Ä–∞–∑–≤–æ—Ä–æ—Ç–∞ –º–∞—Å—Å–∏–≤–∞
# –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –Ω–∞–¥–æ –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–π –º–∞—Å—Å–∏–≤, –∫–æ—Ç–æ—Ä—ã–π –¥–∞—Å—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Ä–∞–∑–Ω–∏—Ü—É curr_rmax - curr_max
# –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã r –∏ l –¥–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –ø–æ—Å–ª–µ —Ç–æ–≥–æ, –∫–∞–∫ –º—ã –Ω–∞–π–¥–µ–º –∫–∞–∫–æ–π –ø–æ–¥–º–∞—Å—Å–∏–≤ —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—å,
# —Ç–æ —Å–º–æ–≥–ª–∏ –±—ã –≤—ã—á–∏—Å–ª–∏—Ç—å –ø—Ä–∞–≤—ã–π –∏ –ª–µ–≤—ã–π –ø–æ–¥–º–∞—Å—Å–∏–≤ –¥–ª—è –æ–∫–æ–Ω—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞

def task8(a, b):
    maxdiff = 0
    result = [a[0] * b[0], 0, 0]
    for mid in range(len(a)):
        # –î–ª—è –Ω–µ—á–µ—Ç–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
        l = mid - 1
        r = mid + 1
        curr_max = a[mid] * b[mid]
        curr_rmax = curr_max
        while l >= 0 and r <= len(a) - 1:
            curr_max += a[l] * b[l] + a[r] * b[r]
            curr_rmax += a[r] * b[l] + a[l] * b[r]
            if curr_rmax - curr_max > maxdiff:
                maxdiff = curr_rmax - curr_max
                result = [curr_rmax, l, r]
            l -= 1
            r += 1

        # –î–ª—è —á–µ—Ç–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
        l = mid
        r = mid + 1
        curr_max = 0
        curr_rmax = 0
        while l >= 0 and r <= len(a) - 1:
            curr_max += a[l] * b[l] + a[r] * b[r]
            curr_rmax += a[r] * b[l] + a[l] * b[r]
            if curr_rmax - curr_max > maxdiff:
                maxdiff = curr_rmax - curr_max
                result = [curr_rmax, l, r]
            l -= 1
            r += 1
    # –ò—â–µ–º —Å—É–º–º—É –ª–µ–≤–æ–≥–æ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    lmax = 0
    for i in range(result[1]):
        lmax += a[i] * b[i]
    # –ò—â–µ–º —Å—É–º–º—É –ø—Ä–∞–≤–æ–≥–æ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    rmax = 0
    for i in range(result[2] + 1, len(a)):
        rmax += a[i] * b[i]
    return lmax + result[0] + rmax

n = int(input())
a = input().split()
a = [int(a[i]) for i in range(len(a))]
b = input().split()
b = [int(b[i]) for i in range(len(b))]
print(task8(a, b))
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –∏—Å–∫–æ–º—ã–π —Ü–µ–ª—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç X, –∏–ª–∏ —Ñ—Ä–∞–∑—É ¬´No solution¬ª, –µ—Å–ª–∏ —Ç–∞–∫–æ–≥–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –∏–ª–∏ –æ–Ω –Ω–µ —Ü–µ–ª—ã–π. –ï—Å–ª–∏ –æ—Ç–≤–µ—Ç–æ–≤ –Ω–µ—Å–∫–æ–ª—å–∫–æ, –≤—ã–≤–µ–¥–∏—Ç–µ –ª—é–±–æ–π.

–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

```python
def task9(A, B, n):
    result = 0
    if A == 0 and B or (A < 0 and B > 0 or B < 0 and A > 0) and n % 2 == 0:
        return "No solution"
    elif A == 0 and B == 0:
        return 5
    elif (A < 0 and B > 0 or B < 0 and A > 0) and n % 2:
        result = -((-B / A) ** (1 / n))
    else:
        result = (B / A) ** (1 / n)
    if result != int(result):
            return "No solution"
    return int(result)

[A, B, n] = input().split()
print(task9(int(A), int(B), int(n)))
```
