```python
import random

#Бинарный поиск
def binary_search(arr, key):
    start = 0
    stop = len(arr) - 1
    while start <= stop:
        mid = start + (stop - start) // 2
        if key == arr[mid]:
            return mid
        elif key < arr[mid]:
            stop = mid - 1
        else:
            start = mid + 1
    return None

def binary_insert(arr, key):
    mid = -1
    start = 0
    stop = len(arr) - 1
    while start <= stop:
        mid = start + (stop - start) // 2
        if key == arr[mid]:
            break
        elif key < arr[mid]:
            stop = mid - 1
        else:
            start = mid + 1 
    if mid != -1 and key > arr[mid]:
        mid = mid + 1
    arr.insert(mid, key)
    return mid

def binary_delete(arr, key):
    index = binary_search(arr, key)
    if index != None:
        del arr[index]
    return index

def binary_rand(length, lbound, hbound):
    arr = []
    for i in range(length):
        binary_insert(arr, random.randint(lbound, hbound))
    return arr

data = binary_rand(10, 10, 99)
print(f"array:                   {data}")
key = random.randint(10, 99)
binary_insert(data, key)
index = binary_search(data, key)
print(f"array with inserted key: {data}")
print(f"inserted key:            {key}")
print(f"key index:               {index}")
binary_delete(data, key)
print(f"array with deleted key:  {data}")
```

```python
import random

#Бинарное дерево
class node:
    def __init__(self, key):
        self.parent = None
        self.left = None
        self.right = None
        self.key = key
    
    def min(self):
        nd = self
        ret = None
        while nd != None:
            ret = nd
            nd = nd.left
        return ret
    
    def max(self):
        nd = self
        ret = None
        while nd != None:
            ret = nd
            nd = nd.right
        return ret
    
    def treewalk(self):
        self.__treewalk(self)
        print()
    
    def __treewalk(self, node):
        if node != None:
            self.__treewalk(node.left)
            print(node.key, end = ' ')
            self.__treewalk(node.right)
    
    def print(self):
        lines, *_ = self.__print()
        for line in lines:
            print(line)

    def __print(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.key
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle
        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left.__print()
            s = '%s' % self.key
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2
        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right.__print()
            s = '%s' % self.key
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2
        # Two children.
        left, n, p, x = self.left.__print()
        right, m, q, y = self.right.__print()
        s = '%s' % self.key
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    
class bst:
    def __init__(self):
        self.root = None
    
    def search(self, key):
        nd = self.root
        while nd != None and key != nd.key:
            if key < nd.key:
                nd = nd.left
            elif key > nd.key:
                nd = nd.right
        return nd
    
    def insert(self, key):
        nd = self.root
        nd_parent = None
        while nd != None:
            nd_parent = nd
            if key < nd.key:
                nd = nd.left
            else:
                nd = nd.right
        if nd_parent == None:
            self.root = node(key)
        elif key < nd_parent.key:
            nd_parent.left = node(key)
            nd_parent.left.parent = nd_parent
        else:
            nd_parent.right = node(key)
            nd_parent.right.parent = nd_parent
    
    def transplant(self, nd1, nd2):
        if nd1.parent == None:
            self.root = nd2
        elif nd1 == nd1.parent.left:
            nd1.parent.left = nd2
        else:
            nd1.parent.right = nd2
        if nd2 != None:
            nd2.parent = nd1.parent
    
    def delete(self, key):
        nd = self.search(key)
        if nd == None:
            return
        elif nd.left == None:
            self.transplant(nd, nd.right)
        elif nd.right == None:
            self.transplant(nd, nd.left)
        else:
            nd1 = nd.right.min()
            if nd1.parent != nd:
                self.transplant(nd1, nd1.right)
                nd1.right = nd.right
                nd1.right.parent = nd1
            self.transplant(nd, nd1)
            nd1.left = nd.left
            nd1.left.parent = nd1
            

def bst_rand(length, lbound, hbound):
    tree = bst()
    for i in range(length):
        tree.insert(random.randint(lbound, hbound))
    return tree

tree = bst_rand(20, 10, 99)
print("tree:")
tree.root.print()
print("tree walk: ", end = '')
tree.root.treewalk()
print("key to delete: ", end = '')
key = int(input())
tree.search(key)
tree.delete(key)
print("tree with deleted key:")
tree.root.print()
print("tree walk with deleted key: ", end = '')
tree.root.treewalk()
```

```python
#Интерполяционный
def interpolation_search(arr, key):
    l = 0
    r = len(arr) - 1
    while arr[l] <= key and arr[r] >= key:
        if l == r:
            if arr[l] == key:
                return l
            return None
        pos = l + (key - arr[l]) * (r - l) // (arr[r] - arr[l])
        if arr[pos] < key:
            l = pos + 1
        elif arr[pos] > key:
            r = pos - 1
        else:
            return pos

def interpolation_insert(arr, key):
    if len(arr) == 0:
        arr.append(key)
        return 0
    l = 0
    r = len(arr) - 1
    while arr[l] <= key and arr[r] >= key:
        if l == r:
            break
        pos = l + (key - arr[l]) * (r - l) // (arr[r] - arr[l])
        if arr[pos] < key:
            l = pos + 1
        elif arr[pos] > key:
            r = pos - 1
        else:
            arr.insert(pos, key)
            return pos
    if arr[l] > key:
        arr.insert(l, key)
        return l
    elif arr[r] < key:
        arr.insert(r + 1, key)
        return r + 1

def interpolation_delete(arr, key):
    index = interpolation_search(arr, key)
    if index != None:
        del arr[index]
    return index

def interpolation_rand(length, lbound, hbound):
    arr = []
    for i in range(length):
        interpolation_insert(arr, random.randint(lbound, hbound))
    return arr

data = interpolation_rand(15, 10, 99)
print(f"array:                   {data}")
key = random.randint(10, 99)
interpolation_insert(data, key)
index = interpolation_search(data, key)
print(f"array with inserted key: {data}")
print(f"inserted key:            {key}")
print(f"key index:               {index}")
interpolation_delete(data, key)
print(f"array with deleted key:  {data}")
```

```python
#Простое рехэширование

def simple_hash(key, sz, probe):
    return (key + probe) % sz

def int_cmp(k1, k2):
    return k1 - k2

#Используем метод открытой адрессации
class openaddr_hashnode:
    def __init__(self):
        self.key = None
        self.deleted = False
class openaddr_hashmap:
    def __init__(self, sz, hfunc, cmp):
        self.table = [openaddr_hashnode() for i in range(sz)]
        self.hash = hfunc
        self.cmp = cmp
    
    def search(self, key):
        for probe in range(len(self.table)):
            node = self.table[self.hash(key, len(self.table), probe)]
            if node.deleted == False and node.key == None:
                break
            elif node.deleted == False and self.cmp(node.key, key) == 0:
                return key
        return None
    
    def insert(self, key):
        for probe in range(len(self.table)):
            node = self.table[self.hash(key, len(self.table), probe)]
            if node.key == None:
                node.key = key
                node.deleted = False
                return key
        return None

    def delete(self, key):
        for probe in range(len(self.table)):
            node = self.table[self.hash(key, len(self.table), probe)]
            if self.cmp(node.key, key) == 0:
                node.key = None
                node.deleted = True
                return key
        return None
    
    def print_table(self):
        print("hash table: ", end = '')
        for i in range(len(self.table)):
            node = self.table[i]
            print(node.key, end = " ")
        print()

def rand_openaddr_hashmap(hm, lbound, hbound):
    rsz = random.randint(1, len(hm.table) - 1)
    for i in range(rsz):
        hm.insert(random.randint(lbound, hbound))
    return hm

hm = openaddr_hashmap(13, simple_hash, int_cmp)
rand_openaddr_hashmap(hm, 10, 99)
hm.print_table()
print("key to delete: ", end = '')
key = int(input())
hm.delete(key)
hm.print_table()
print("key to insert: ", end = '')
key = int(input())
hm.insert(key)
hm.print_table()
```

```python
#Рехэширование с помощью псевдослучайных чисел
def probing_sequence(length):
    arr = [i for i in range(length)]
    for i in range(len(arr)):
        index = random.randint(i, len(arr) - 1)
        temp = arr[i]
        arr[i] = arr[index]
        arr[index] = temp
    return arr

arrlen = 13
sequence = probing_sequence(arrlen)
def rand_hash(key, sz, probe):
    return (key + sequence[probe]) % sz

hm = openaddr_hashmap(arrlen, rand_hash, int_cmp)
rand_openaddr_hashmap(hm, 10, 99)
hm.print_table()
print("key to delete: ", end = '')
key = int(input())
hm.delete(key)
hm.print_table()
print("key to insert: ", end = '')
key = int(input())
hm.insert(key)
hm.print_table()
```

```python
#Метод цепочек
def hashf(key, sz):
    return key % sz

class chain_hashmap:
    def __init__(self, sz, hfunc, cmp):
        self.table = [[] for i in range(sz)]
        self.hash = hfunc
        self.cmp = cmp
    
    def search(self, key):
        bucket = self.table[self.hash(key, len(self.table))]
        for i in range(len(bucket)):
            if self.cmp(bucket[i], key) == 0:
                return bucket[i]
        return None
    
    def insert(self, key):
        bucket = self.table[self.hash(key, len(self.table))]
        bucket.append(key)
        return key

    def delete(self, key):
        bucket = self.table[self.hash(key, len(self.table))]
        for i in range(len(bucket)):
            if bucket[i] == key:
                del bucket[i]
                return key
        return None
    
    def print_table(self):
        print("hashmap:")
        for i in range(len(self.table)):
            bucket = self.table[i]
            print(f"    bucket {i}: ", end = '')
            for j in range(len(bucket)):
                print(bucket[j], end = ' ')
            print()

def rand_chain_hashmap(hm, lbound, hbound):
    rsz = random.randint(1, len(hm.table) - 1)
    for i in range(rsz):
        hm.insert(random.randint(lbound, hbound))
    return hm
    
hm = chain_hashmap(7, hashf, int_cmp)
rand_chain_hashmap(hm, 10, 99)
hm.print_table()
print("key to delete: ", end = '')
key = int(input())
hm.delete(key)
hm.print_table()
print("key to insert: ", end = '')
key = int(input())
hm.insert(key)
hm.print_table()
```

Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу, которая находит хотя бы один способ решения задач.


```python
def place_queen(board, i, j):
    for k in range(len(board)):
        board[i][k] += 1
        board[k][j] += 1
        if 0 <= i + j - k < len(board):
            board[i + j - k][k] += 1
        if 0 <= i - j + k < len(board):
            board[i - j + k][k] += 1
    board[i][j] = -1

def remove_queen(board, i, j):
    for k in range(len(board)):
        board[i][k] -= 1
        board[k][j] -= 1
        if 0 <= i + j - k < len(board):
            board[i + j - k][k] -= 1
        if 0 <= i - j + k < len(board):
            board[i - j + k][k] -= 1
    board[i][j] = 0

def print_answer(board):
    res = ""
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] == -1:
                res += chr(97 + i) + str(j + 1) + ' '
    print(res)
                

def __queens_problem(board, j):
    if j == len(board):
        print_answer(board)
        return
    for i in range(len(board)):
        if board[i][j] == 0:
            place_queen(board, i, j)
            __queens_problem(board, j + 1)
            remove_queen(board, i, j)

def queens_problem(n):
    board = [[0 for i in range(n)] for j in range(n)]
    __queens_problem(board, 0)
    
queens_problem(8)
```
