### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit

def print_matrix(mat):
    print()
    for i in range(user_m):
        for j in range(user_n):
            print(mat[i][j], end = ' ')
        print()

matrix = []
user_m = int(user_m)
user_n = int(user_n)
user_min_limit = int(user_min_limit)
user_max_limit = int(user_max_limit)
for i in range(user_m):
    matrix.append([0] * user_n)
for i in range(user_m):
    for j in range(user_n):
        matrix[i][j] = random.randint(user_min_limit, user_max_limit)

print_matrix(matrix)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.

mat = copy.deepcopy(matrix)

start_time = time.time()

def selection_sort(arr):
    for i in range(len(arr)):
        index = i
        for j in range(i + 1, len(arr)):
            if arr[index] > arr[j]:
                index = j
        temp = arr[index]
        arr[index] = arr[i]
        arr[i] = temp
            

for i in range(len(mat)):
    selection_sort(mat[i])

selection_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(selection_sort_time))

print_matrix(mat)
```

```python
# Сортировка вставкой.

mat = copy.deepcopy(matrix)

start_time = time.time()

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        
for i in range(len(mat)):
    insertion_sort(mat[i])

insertion_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(insertion_sort_time))

print_matrix(mat)
```

```python
# Сортировка обменом. || Сортировка пузырьком.

mat = copy.deepcopy(matrix)

start_time = time.time()

def bubble_sort(arr):
    for i in range(len(arr)):
        swap_cnt = 0
        for j in range(1, len(arr) - i):
            if arr[j] > arr[j - 1]:
                temp = arr[j]
                arr[j] = arr[j - 1]
                arr[j - 1] = temp
                swap_cnt += 1
        if swap_cnt == 0:
            break
            
        
for i in range(len(mat)):
    bubble_sort(mat[i])

bubble_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(bubble_sort_time))


print_matrix(mat)
```

```python
# Сортировка Шелла.

mat = copy.deepcopy(matrix)

start_time = time.time()

def shell_sort(arr):
    offset = int(len(arr) / 2)
    while offset > 0:
        for i in range(offset, len(arr), offset):
            key = arr[i]
            j = i - offset
            while j >= 0 and key < arr[j]:
                arr[j + offset] = arr[j]
                j -= offset
            arr[j + offset] = key
        offset = int(offset / 2)
        
for i in range(len(mat)):
    shell_sort(mat[i])
        

shell_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(shell_sort_time))

print_matrix(mat)
```

```python
# Быстрая сортировка.

mat = copy.deepcopy(matrix)

start_time = time.time()

def partition(arr, start, end):
    pivot = arr[start]
    left = start + 1
    right = end
    while True:
        while arr[right] > pivot and right > start:
            right -= 1
        while arr[left] <= pivot and left < end:
            left += 1
        if left >= right:
            break
        temp = arr[left]
        arr[left] = arr[right]
        arr[right] = temp
    arr[start] = arr[right]
    arr[right] = pivot
    return right

def __quick_sort(arr, start, end):
    if end > start:
        pivot_index = partition(arr, start, end)
        __quick_sort(arr, start, pivot_index - 1)
        __quick_sort(arr, pivot_index + 1, end)
   

def quick_sort(arr):
    __quick_sort(arr, 0, len(arr) - 1)
        
        
for i in range(len(mat)):
    quick_sort(mat[i])
        

quick_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(quick_sort_time))

print_matrix(mat)
```

```python
# Турнирная сортировка.

import heapq

mat = copy.deepcopy(matrix)

start_time = time.time()
    
def tournament_sort(arr):
    heap = arr.copy() 
    winners = []
    losers = []
    winners.append(heapq.heappop(heap))
    for i in range(len(heap)):
        heap_top = heapq.heappop(heap)
        if heap_top >= winners[-1]:
            winners.append(heap_top)
        else:
            losers.append(heap_top)
    arr.clear()
    arr.extend(losers)
    arr.extend(winners)
    if len(losers) > 0:
        tournament_sort(arr)
    
    
for i in range(len(mat)):
    tournament_sort(mat[i])

tournament_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(tournament_sort_time))

print_matrix(mat)
```
