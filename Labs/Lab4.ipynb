Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
#Stack
class stack:
    def __init__(self):
        self.mem = []
    
    def push(self, key):
        self.mem.append(key)
    
    def pop(self):
        if self.isempty():
            return None
        key = self.mem[-1]
        del self.mem[-1]
        return key
    
    def peek(self):
        return self.mem[-1]
    
    def isempty(self):
        if len(self.mem):
            return False
        return True
```

```python
#Deque
class deque:
    def __init__(self):
        self.mem = []
        
    def pushback(self, key):
        self.mem.append(key)
        
    def pushfront(self, key):
        self.mem.insert(0, key)
        
    def popback(self):
        if self.isempty():
            return None
        key = self.mem[-1]
        del self.mem[-1]
        return key
        
    def popfront(self):
        if self.isempty():
            return None
        key = self.mem[0]
        del self.mem[0]
        return key
    
    def peekback(self):
        return self.mem[-1]
    
    def peekfront(self):
        return self.mem[0]

    def isempty(self):
        if len(self.mem):
            return False
        return True
```

```python
#Read file to list
def readfile(filename):
    with open(filename) as file:
        return file.readlines()

def readchars(filename):
    with open(filename) as file:
        return list(file.read())
    
#Write list to file
def writefile(filename, text):
    with open(filename, 'w') as file:
        for str in text:
            file.write(str)
    
#Print array of lines
def printlines(text):
    for elem in text:
        print(elem, end = '')
        
def print_chars(arr):
    for item in arr:
        print(item, end = '')
    print()
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
def deque_sort(arr):
    open = deque()
    open.mem = arr.copy()
    closed = deque()
    while not open.isempty():
        min = open.peekback()
        for i in open.mem:
            if open.peekback() < min:
                min = open.peekback()
            open.pushfront(open.popback())
        while open.peekback() != min:
            open.pushfront(open.popback())
        closed.pushback(open.popback())
    return closed.mem

strs = readfile("lab4_files/1in.txt")
print("books:")
printlines(strs)
print("\nsorted books:")
printlines(deque_sort(strs))
```

### Задание №2
Дек содержит последовательность символов для шифровки сообщений. Дан текстовый файл, содержащий зашифрованное сообщение. Пользуясь деком, расшифровать текст. 
Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
def decrypt_message(encrypted_msg, alphabet_mem):
    alphabet = deque()
    alphabet.mem = alphabet_mem.copy()
    decrypted_msg = ""
    for char in encrypted_msg:
        for i in alphabet.mem:
            if alphabet.peekback() != char:
                alphabet.pushfront(alphabet.popback())
            else:
                alphabet.pushback(alphabet.popfront())
                alphabet.pushback(alphabet.popfront())
                decrypted_msg += alphabet.peekback()
                break
    return decrypted_msg

alphabet = "lrd! oHew"
print(f"deque data: {alphabet}")
msg = readchars("lab4_files/2in.txt")
print(f"encrypted msg: ", end = '')
print_chars(msg)
print(f"decrypted msg: {decrypt_message(msg, list(alphabet))}")
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
class tower:
    def __init__(self, name):
        self.name = name
        self.stack = stack()

def __hanoi_tower(len, A, B, C):
    if len:
        __hanoi_tower(len - 1, A, C, B)
        key = A.stack.pop()
        C.stack.push(key)
        print(f"move {key} from {A.name} to {C.name}")
        __hanoi_tower(len - 1, B, A, C)

def hanoi_tower(A):
    A_tower = tower("A")
    A_tower.stack.mem = A.copy()
    B_tower = tower("B")
    C_tower = tower("C")
    __hanoi_tower(len(A_tower.stack.mem), A_tower, B_tower, C_tower)
    return C_tower.stack.mem

A = readfile('lab4_files/3in.txt')
A = [int(item) for item in A]
print(f'A tower: {A}')
C = hanoi_tower(A)
print(f'C tower: {C}')
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def check_crc_balance(text):
    balance = 0
    s = stack()
    s.mem = text.copy()
    while not s.isempty():
        key = s.pop()
        if key == '(':
            balance += 1
        elif key == ')':
            balance -= 1
        if balance > 0:
            return False
    if balance:
        return False
    return True

text = readchars('lab4_files/4in.txt')
print_chars(text)
print(f'is balanced: {check_crc_balance(text)}')
```

### Задание №5
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс квадратных скобок в тексте, используя дек.

```python
def check_sqr_balance(text):
    balance = 0
    d = deque()
    d.mem = text.copy()
    while not d.isempty():
        key = d.popfront()
        if key == '[':
            balance += 1
        elif key == ']':
            balance -= 1
        if balance < 0:
            return False
    if balance:
        return False
    return True

text = readchars('lab4_files/5in.txt')
print_chars(text)
print(f'is balanced: {check_sqr_balance(text)}')
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
def stack_chars(arr):
    digit = []
    alpha = []
    other = []
    s = stack()
    s.mem = arr.copy()
    s.mem.reverse()
    while not s.isempty():
        key = s.pop()
        if key.isdigit():
            digit.append(key)
        elif key.isalpha():
            alpha.append(key)
        else:
            other.append(key)
    return [digit, alpha, other]

text = readchars('lab4_files/6in.txt')
arr = stack_chars(text)
print_chars(text)
print_chars(arr[0])
print_chars(arr[1])
print_chars(arr[2])
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
def deque_negpos(arr):
    res_neg = []
    res_pos = []
    d1 = deque()
    d1.mem = arr.copy()
    while not d1.isempty():
        key = d1.popback()
        if key >= 0:
            res_pos.insert(0, key)
        else:
            res_neg.insert(0, key)
    return [res_neg, res_pos]

arr = readfile('lab4_files/7in.txt')
arr = [int(item) for item in arr]
print(f'Все числа: {arr}')
res = deque_negpos(arr)
print(f'Отрицательные числа: {res[0]}')
print(f'Положительные числа: {res[1]}')
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
def stack_reverse(arr):
    st1 = stack()
    st1.mem = arr
    st2 = stack()
    while not st1.isempty():
        st2.push(st1.pop())
    return st2.mem

text = readfile('lab4_files/8in.txt')
print('Исходный текст:')
printlines(text)
text = stack_reverse(text)
print('\nПеревернутый текст:')
printlines(text)
writefile('lab4_files/8out.txt', text)
```
