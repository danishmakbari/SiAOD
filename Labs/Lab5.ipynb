### Выполнение:

```python
import turtle
import time
from collections import defaultdict

def tsetup():
    window = turtle.Screen()
    window.setup(800, 800)
    window.title("Фракталы")
    window.tracer(0)
    turtle.reset()
    turtle.setpos(0, 0)
    turtle.hideturtle()
    turtle.clear()
    return window
    
def tclose(window):
    window.update()
    window.mainloop()
    
def testing(frac_func, depth_min, depth_max):
    test_num = 1
    print("Таблица тестов:\nTest№ Depth Time")
    for i in range(depth_min, depth_max + 1):
        tstart = time.time()
        frac_func(i)
        tend = time.time() - tstart
        print(f"{test_num}     {i}     {tend}")
        test_num += 1

def lsystem(axiom, rules, depth):
    for i in range(depth):
        temp = ""
        for char in axiom:
            temp += rules[char]
        axiom = temp
    return axiom
```

```python
# Кривая Коха

def lsystem_koch_curve(depth):
    rules = defaultdict()
    rules['F'] = "F-F++F-F"
    rules['+'] = '+'
    rules['-'] = '-'
    return lsystem('F', rules, depth)

def __koch_curve(length, depth):
    axiom = lsystem_koch_curve(depth)
    for char in axiom:
        if char == 'F':
            turtle.forward(length)
        elif char == '+':
            turtle.right(60)
        else:
            turtle.left(60)

def koch_curve(length, depth):
    try:
        window = tsetup()
        __koch_curve(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __koch_curve(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
koch_curve(2, depth)
testing(lsystem_koch_curve, 0, 7)
```

```python
# Кривая Гильберта

def lsystem_hilbert_curve(depth):
    rules = defaultdict()
    rules['F'] = 'F'
    rules['+'] = '+'
    rules['-'] = '-'
    rules['A'] = "+BF-AFA-FB+"
    rules['B'] = "-AF+BFB+FA-"
    return lsystem('A', rules, depth)

def __hilbert_curve(length, depth):
    axiom = lsystem_hilbert_curve(depth)
    for char in axiom:
        if char == 'F':
            turtle.forward(length)
        elif char == '+':
            turtle.left(90)
        elif char == '-':
            turtle.right(90)

def hilbert_curve(length, depth):
    try:
        window = tsetup()
        __hilbert_curve(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __hilbert_curve(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
hilbert_curve(5, depth)
testing(lsystem_hilbert_curve, 0, 7)
```

```python
# Кривая Серпинского

def lsystem_serpinski_curve(depth):
    rules = defaultdict()
    rules['+'] = '+'
    rules['-'] = '-'
    rules['F'] = 'F'
    rules['G'] = 'G'
    rules['X'] = "XF+G+XF--F--XF+G+X"
    return lsystem("F--XF--F--XF", rules, depth)

def __serpinski_curve(length, depth):
    axiom = lsystem_serpinski_curve(depth)
    for char in axiom:
        if char == 'F' or char == 'G':
            turtle.forward(length)
        elif char == '+':
            turtle.left(45)
        elif char == '-':
            turtle.right(45)

def serpinski_curve(length, depth):
    try:
        window = tsetup()
        __serpinski_curve(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __serpinski_curve(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
serpinski_curve(5, depth)
testing(lsystem_serpinski_curve, 0, 7)
```

```python
# Салфетка Серпинского

def lsystem_serpinski_triangle(depth):
    rules = defaultdict()
    rules['+'] = '+'
    rules['-'] = '-'
    rules['A'] = "B-A-B"
    rules['B'] = "A+B+A"
    return lsystem('A', rules, depth)

def __serpinski_triangle(length, depth):
    axiom = lsystem_serpinski_triangle(depth)
    for char in axiom:
        if char == 'A' or char == 'B':
            turtle.forward(length)
        elif char == '+':
            turtle.left(60)
        elif char == '-':
            turtle.right(60)

def serpinski_triangle(length, depth):
    try:
        window = tsetup()
        __serpinski_triangle(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __serpinski_triangle(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
serpinski_triangle(5, depth)
testing(lsystem_serpinski_triangle, 0, 7)
```

```python
# Ковер Серпинского

def lsystem_serpinski_carpet(depth):
    rules = defaultdict()
    rules['+'] = '+'
    rules['-'] = '-'
    rules['F'] = "F+F-F-F-G+F+F+F-F"
    rules['G'] = "GGG"
    return lsystem('F', rules, depth)

def __serpinski_carpet(length, depth):
    axiom = lsystem_serpinski_carpet(depth)
    for char in axiom:
        if char == 'F':
            turtle.forward(length)
        elif char == 'G':
            turtle.penup()
            turtle.forward(length)
            turtle.pendown()
        elif char == '+':
            turtle.right(90)
        elif char == '-':
            turtle.left(90)

def serpinski_carpet(length, depth):
    try:
        window = tsetup()
        __serpinski_carpet(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __serpinski_carpet(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
serpinski_carpet(5, depth)
testing(lsystem_serpinski_carpet, 0, 7)
```

```python
# Дерево Пифагора

def lsystem_pythagoras_tree(depth):
    rules = defaultdict()
    rules['+'] = '+'
    rules['-'] = '-'
    rules['['] = '['
    rules[']'] = ']'
    rules['1'] = "11"
    rules['0'] = "1[0]0"
    return lsystem('0', rules, depth)

def __pythagoras_tree(length, depth):
    axiom = lsystem_pythagoras_tree(depth)
    turtle.left(90)
    stack = []
    for char in axiom:
        if char == '0' or char == '1':
            turtle.forward(length)
        elif char == '[':
            stack.append(turtle.heading())
            stack.append(turtle.pos())
            turtle.left(45)
        elif char == ']':
            pos = stack.pop()
            turtle.penup()
            turtle.setpos(pos[0], pos[1])
            turtle.pendown()
            turtle.setheading(stack.pop())
            turtle.right(45)

def pythagoras_tree(length, depth):
    try:
        window = tsetup()
        __pythagoras_tree(length, depth)
        tclose(window)
    except:
        window = tsetup()
        __pythagoras_tree(length, depth)
        tclose(window)

print("Глубина: ", end = '')
depth = int(input())
pythagoras_tree(2, depth)
testing(lsystem_pythagoras_tree, 0, 7)
```
